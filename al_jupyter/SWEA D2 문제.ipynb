```python
T = int(input())
for tc in range(1, T+1):
    word = input()
    
    n = len(word)
    i = 0
    rlt = 0
    while i != int(n/2):
        if word[i+0] == word[-1-i]:
            rlt = 1
        else:
            rlt = 0
            break
        
        i += 1
        
    print(f'#{tc} {rlt}')
```
