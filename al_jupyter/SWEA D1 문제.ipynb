```python
# 평균값 = 모든 더한 값 / 개수

T = int(input())

for tc in range(1, T+1):
    lst = list(map(int, input().split()))
    lst_sum = 0
    
    
    for i in lst:
        lst_sum += i
    
    # 반올림해서 정수여야 함
    rlt = round(lst_sum / len(lst))
    
    print(f'#{tc} {rlt}')
```

```python
T = int(input())

for tc in range(1, T+1):
    lst = list(map(int, input().split()))
    
    max_number = lst[0]
    
    for i in range(len(lst)):
        if max_number < lst[i]:
            max_number = lst[i]
            
    print(f'#{tc} {max_number}')
```

```python
N = int(input())
lst = list(map(int, input().split()))

# 정렬
for i in range(N):
    for j in range(i):
        if lst[i] >= lst[j]:
            lst[i], lst[j]  = lst[j], lst[i]


# 중간 값 
mid_number = lst[int(N/2)]

print(f'{mid_number}')
```

```python
# 문자열 메소드 안쓰고 ord() chr()로 풀어보기
n = input()
#print(n[0][0]) # T
#print(chr(ord(n[0][0])+32))
new = ''
for i in n:
    if 97 <= ord(i) <= 122:
        new += chr(ord(i) -32)
    elif 65 <= ord(i) <= 90:
        new += i
    else:
        new += i
print(new)
```
