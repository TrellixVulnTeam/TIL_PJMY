# 카운트 소트 : DAT


```python
# DAT (Direct Address Table)
# 출력값 : O X O O X X O X O

a = [4, 9, 1, 4, 4, 2] # n개의 정수
b = [1, 3, 4, 2, 5, 6, 7, 8, 9] # m개의 정수

for i in range(len(b)):
    flag = ''
    for j in range(len(a)):
        if a[j] == b[i]:
            flag = '같은 값이 있음'
            print(flag, a[j])
            break
```

```python
# 안보고 짜보기

a = [4, 9, 1, 4, 4, 2] # n개의 정수
b = [1, 3, 4, 2, 5, 6, 7, 8, 9] # m개의 정수

# flag의 위치에 따라 달라지는 값..

flag = 0
for i in b:    
    for j in a:
        if i == j:
            flag +=1
            print(flag, i)
```

```python
'''
DAT  direct address table 자료구조..
빠른 검색을 위한 자료구조  dat
count sort 기본 원리


[의사코드]
1. bucket을 생성함.
2. 일단 길이가 작은 리스트를 bucket에 등록.
3. bucket[a[i]] 이런식으로 쓰는 것.
4. 
'''

# 출력값: O X O O X X O X O
# B 배열 안에 있는 값이 A에 있으면 O 출력!!

a=[4,9,1,4,4,2]
b=[1,3,4,2,5,6,7,8,9]

bucket = [0]*10
for i in range (len(a)):
    index=a[i]
    bucket[index]=1

print(bucket)
print('------------------------')


for i in range (len(b)):
    if bucket[b[i]] == 1:
        print('O',end=' ')
    else:
        print('X',end=' ')
```

```python
# 함수 내부에 불필요한 print문이 있는 경우 오답으로 처리가 됩니다.
from defusedxml import DTDForbidden


def is_position_safe(N, M, position):
    # N * N 의 제한 구역이 있음
    # 1. 각 모서리에서 제한 범위를 빠져나가는 경우의 수를 나눔
    # 2. 모서리를 제외한 가장자리도 나눔
    
    # position == (0, 0) 일 때 
    if position == (0, 0):
        if M == 0 or M == 2:
            return False

    # position == (0, N) 일 때
    if position == (0, N):
        if M == 0 or M == 3:
            return False
    
    # position == (N, 0) 일 때 
    if position == (N, 0):
        if M == 2 or M == 1:
            return False

    # position == (N, N) 일 때         
    if position == (N, N):
        if M == 1 or M == 3:
            return False
    

    # position == (1 ~ N-1, N) 일 때 
    for i in range(1, N):
        if position == (i, N):
            if M == 3:
                return False


    # position == (0, 1 ~ N-1) 일 때 
    for i in range(1, N):
        if position == (0, i):
            if M == 0:
                return False

    # position == (1 ~ N-1, 0) 일 때 
    for i in range(1, N):
        if position == (i, 0):
            if M == 2:
                return False

    # position == (N, 1 ~ N-1) 일 때 
    for i in range(1, N):
        if position == (N, i):
            if M == 1:
                return False
    
    return True 
    
        


# 아래의 코드를 수정하거나 새롭게 추가하지 않습니다.
if __name__ == '__main__':
    print(is_position_safe(3, 1, (0, 0))) # True
    print(is_position_safe(3, 0, (0, 0))) # False
    
```

```python
# 여러개가 있다면 여러개를 다 출력함.




#pattern = [list(map(str, input())) for _ in range(2)]

def is_pattern(m):
    # pattern 입력받기
    pattern = []
    rlt = ''
    for row in range(2):
        row = list(map(str, input())) 
        pattern.append(row)

    # a = m[0][2:4]
    # print(a)    
    # m 배열에 pattern이 존재하는지 확인하기 - 1개가 발견 되었다면 발견(1개), 없다면 미발견
    cnt = 0
    for i in m: # 헹 
        for j in range(len(m[0])-1):
            if i[j:j+2] == pattern[0]:
                #print(f'발견: {pattern[0]}')
                cnt += 1
            if i[j:j+2] == pattern[1]:
                cnt += 1

    if cnt > 0:
        rlt = f"발견({cnt}개)"
    else:
        rlt = "미발견"

    return rlt


map_1 = [['A', 'B', 'G', 'K'], ['T', 'T', 'A', 'B'], ['A', 'C', 'C', 'D']]
ans = is_pattern(map_1)
print(f'"{ans}"')
```
