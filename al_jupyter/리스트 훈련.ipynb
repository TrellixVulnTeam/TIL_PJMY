* 전치행렬


```python
# 2. 전치 행렬 뽑기 (for문 사용)
'''
전치행렬 : 대각선 기준으로 바뀜
'''

import pprint

a = [[4, 5, 6, 7, 8], [4, 7, 8, 9, 2], [3, 1, 6, 8, 9], [3, 1, 8, 9, 0], [3, 1, 9, 6, 7]]
pprint.pprint(a)
print('-------------------------------------')

for i in range(len(a)):
    for j in range(len(a[0])):
        if i == j :
            continue
        elif i > j:
            a[i][j], a[j][i] = a[j][i], a[i][j]
            
pprint.pprint(a)
```

```python
# 2-1. 열로 뽑기
import pprint

a = [[4, 5, 6, 7, 8], [4, 7, 8, 9, 2], [3, 1, 6, 8, 9], [3, 1, 8, 9, 0], [3, 1, 9, 6, 7]]
b = list(map(list, zip(*a)))  # 이러면 리스트
pprint.pprint(b)
```
