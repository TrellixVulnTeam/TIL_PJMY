```python
a, b, c = map(int, input().split())
rlt = ''

if a == b and a == c and b == c:
    rlt = '모두 같다'

elif a != b and a != c and b != c:
    rlt = '모두 다르다'
else:
    rlt = '일부 같다'
print(rlt)
```

```python
n = int(input())

lst = []
cnt = 0
for i in range(1, n+1):
    new_i = str(i)
    if ('3' in new_i) or  ('6' in new_i) or ('9' in new_i):
        
        for j in new_i:
            if '3' == j:
                cnt+= 1
            elif '6' == j:
                cnt+= 1
            elif '9' == j:
                cnt+= 1
        lst.append('짝' * cnt)
        cnt = 0
    else:
        lst.append(i)
        
        
print(*lst)
```
