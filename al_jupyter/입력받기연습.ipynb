```python
import pprint
board = [[0 for _ in range(5)] for _ in range(5)]

for i in range(len(board)): # 행의 값 = max는 4
    for j in range(len(board[0])): # 열의 값 = max는 5
        if i >= j: # 열이 행보다 작을 때 
            board[i][j] = i+1

pprint.pprint(board)
```

# 2차원 배열의 입출력

```python
'''
[입력]
4 5
1 0 0 0 0
2 2 0 0 0
3 3 3 0 0
4 4 4 4 0
'''

row, col = map(int, input().split())
board = [list(map(int, input().split())) for _ in range(row)]

# 출력
print('--------아래는 결과값-----------')
for i in range(row):
    for j in range(col):
        print(board[i][j], end=' ') # 한행 다 출력
    print() # 다음행 출력 전 \n 출력
```

```python
'''
이 문제는 벽을 만날 때 까지, 혹은 경계면을 만날때까지 십자가를 출력하는 문제였음
5 5         # 5 x 5 라는 뜻
0 0 0 0 0   # 다섯줄을 받아서 board를 만듦
0 0 0 2 0 
0 0 0 0 0
2 2 2 0 0 
0 0 0 0 0
1 1         # 시작값이라는 뜻
'''

# 함수를 만들어 볼 거야 ~

def cross():
    if board[sr][sc] != 0:  # 예제 코드에서는 0 또는 2니까, 0이 아니면 2(벽)임.
        return
    board[sr][sc] = 1 # 이것은 시작 위치를 잡는 것 : 이 위치부터 1로 마킹
    
    for i in range(sr-1, -1, -1): # 하나씩 감소하면서 위로 올라감
        if board[i][sc] != 0:
            break
        board[i][sc] = 1
    
    for i in range(sr+1, row): # 하나씩 증가하면서 아래로 내려감
        if board[i][sc] != 0:
            break
        board[i][sc] = 1
    
    for i in range(sc-1, -1, -1):
        if board[sr][i] != 0:
            break
        board[sr][i] = 1
    
    for i in range(sc+1, col):
        if board[sr][i] != 0:
            break
        board[sr][i] = 1
    
    
    #return board : 


row, col = map(int, input().split()) # 행, 열을 받음

# 2차원 리스트를 입력받음
board = [list(map(int, input().split())) for _ in range(row)]

# 시작 위치
sr, sc = map(int, input().split())

cross()
print('--------아래는 결과값-----------')
for i in range(row):
    for j in range(col):
        print(board[i][j], end=' ') # 한행 다 출력
    print() # 다음행 출력 전 \n 출력
```
