* 버블정렬은 인접한 두개의 원소를 비교하며 자리를 계속 교환하는 방식이다.

[절대값 차 구하기]


```python
# 내가 만든 것
from pprint import pprint
arr = []
for i in range(5):
    arr.append([j+ i* 5 for j in range(1, 6)])
# pprint(arr)

rlt = 0

# 각 인덱스에 접근
for i in range(5):
    for j in range(5):
        d_row = [-1, 1, 0, 0] # 상 하 좌 우
        d_col = [0, 0, -1, 1]
        for d in range(4):    
            ii = i +d_row[d]
            jj = j +d_col[d]
            
            if ii < 0 or ii > 4 or jj < 0 or jj >4:
                continue
            else:
                sub = arr[i][j] - arr[ii][jj]
                if sub < 0 :
                    sub = -1 * sub
                
                rlt+= sub
        
        
print(rlt)
```

```python
# 1. matrix 초기화
matrix = []
for i in range(5):
    row = [j + i * 5 for j in range(1, 6)]
    matrix.append(row)
    #print(row)

rlt = 0
# 2. 각 요소에 접근하기
for r in range(len(matrix)):
    for c in range(len(matrix[0])):
        d_row = [-1, 1, 0, 0] # 상 하 좌 우
        d_col = [0, 0, -1, 1]
        for d in range(4):
            each_row = r + d_row[d]
            each_col = c + d_col[d]
            # 범위 체크 : 없는 값에 접근할 때 index error가 남
            if each_row < 0 or each_row > 4 or each_col < 0 or each_col > 4:
                continue
            else:
                # print(f'target{matrix[r][c]} --> {matrix[each_row][each_col]}')
                # 3. 각각의 요소의 차 구하기
                sub = matrix[r][c] - matrix[each_row][each_col]
                # 4. 절대값 처리하기
                if sub < 0 :
                    sub = -1 * sub
                # 삼항 연산자 sub = sub if sub >= 0 else -sub

                # 5. 합 누적하기기
                rlt += sub

print(rlt)
```

```python
def bi(a, low, high, key):
    if low > high:
        return False
    else:
        middle = (low+high) //2
        if key == a[middle]:
            return True
        elif key < a[middle]:
            return bi(a, low, middle-1, key)
        elif key > a[middle]:
            return bi(a, middle+1, high, key)

arr = [1, 3, 4, 6, 7, 8, 9, 14, 15, 16]        
start = 1
end = 16
bi(arr, start, end, 10)
```
