```python
N = int(input())

for i in range(1, N+1):
    if N % i == 0:
        print(f'{i}(은)는 {N}의 약수입니다.')
```

```python
N = int(input())
count = []
for i in range(1, N+1):
    if N % i == 0:
        count.append(i)

if len(count) == 2:
    print(f'{count[0]}(은)는 {N}의 약수입니다.')
    print(f'{count[1]}(은)는 {N}의 약수입니다.')
    print(f'{N}(은)는 {count[0]}과 {count[1]}로만 나눌 수 있는 소수입니다.')
```

```python
T = input()
ord_T = ord(T)

if 97 <= ord_T <= 122:
    print(f'{T}는 소문자입니다.')
```
