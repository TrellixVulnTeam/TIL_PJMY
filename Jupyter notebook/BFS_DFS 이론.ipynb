# Graph

1. 인접 행렬
- 2차원 배열을 이용해서 간선 정보를 저장

2. 인접 리스트
- 정점에서 해당 정점으로 나가는 간선의 정보를 저장

3. 간선의 배열
- 간선(시작 정점, 끝 정점)을 배열에 연속적으로 저장

# 그래프 순회
- 그래프 순회는 비선형구조인 그래프로 표현된 모든 자료(정점)을 빠짐없이 탐색하는 것을 의미

1. 깊이 우선 탐색 (Depth First Search, DFS)
- 후입선출 구조의 스택 사용


2. 너비 우선 탐색 (Breadth First Search, BFS)
- 선입선출 구조의 큐를 사용


- 큐의 연산 과정
