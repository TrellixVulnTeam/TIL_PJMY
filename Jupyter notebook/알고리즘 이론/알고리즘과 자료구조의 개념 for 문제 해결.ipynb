### 2-1) 배열

> 배열에서 주로 하는 일
* 삽입(값을 담고), 삭제(삭제하고), 탐색(원하는 값을 찾음) 연산

> 삽입 연산 O(N)
* 배열이 주어졌을 때, 앞이나 가운데에 값을 넣는다고 가정합시다. 

* 새로운 값이 들어갈 자리를 확보하기 위해 다른 값들이 한칸씩 이동해야 합니다. 

* 따라서 일반적으로 삽입의 시간복잡도는 O(N)

* 항상 배열의 맨 뒤에 새로운 값을 삽입하는 경우만 발생한다면, 이런 경우엔 다른 값들의 이동이 전혀 존재하지 않기 때문에 O(1) 의 시간복잡도를 보일 것입니다.

> 삭제 연산 O(N)
* 맨 뒤의 값을 제거하게 된다면 O(1)

* 그러나 삭제의 주 목적은 우리가 원하는 값을 지우는 것이기 때문에,
맨 뒤의 값이 아닌 다른 값을 제거하는 경우도 살펴봐야 할 것입니다.

* 최악의 경우, 맨 앞의 값을 삭제하게 된다면 나머지 N-1개의 값들이 모두 이동해야 하기 때문에 O(N)의 시간복잡도

> 탐색 연산 O(N)
* 자연스럽게 탐색의 시간복잡도도 마찬가지로 O(N)

> k번째 원소 값 구하기 O(1)
* 배열이 주어졌을 때, k번째 원소를 찾는 데 걸리는 시간은,
배열은 index 기반으로 이루어져 있기 때문에 k - 1번째 index를 참조하면 바로 k번째 원소를 구할 수 있습니다.

### 2-2) 동적 배열

> Python에서는 리스트가 곧 동적 배열
* Python에서는 정적으로 배열 선언할 수 있는 방법이 없음
* 동적 배열은 자유롭게 길이가 줄어들고 늘어날 수 있음
* 정확히 사용하고 싶은 만큼만 공간을 차지하여 사용하는 방식

> 동적 배열의 삽입 삭제 탐색 과정 = 정적 배열과 동일
* 시간복잡도가 완전히 일치하지만, 메모리 차이가 있음
* 크기가 고정되어 있는 문제의 경우에는 정적 배열과 같은 양의 데이터를 저장해도
정적 배열보다는 더 많은 메모리를 사용하게 됩니다.

> 동적 배열 각 함수 구현



```python
# 입력
'''
9
push_back 10
push_back 20
get 1
get 2
size
pop_back
size
get 1
size
'''

# 출력
'''
10
20
2
1
10
1
'''
```

```python
# 변수 선언 및 입력
n = int(input())

# 동적 배열 선언
v = list()

# 
for _ in range(n):   
    command = input()

    if command.startswith("push_back"):
        _, num = tuple(command.split())
        v.append(int(num))
    elif command.startswith("pop_back"):
        v.pop()
    elif command.startswith("size"):
        print(len(v))
    else:
        _, index = tuple(command.split()) # get 1 들어오면 1번째 원소
        print(v[int(index) - 1])          # index는 0부터니까

# 
```

### 2-3) 단일 연결 리스트

> 배열은 중간 삽입과 삭제의 시간복잡도가 O(N)
* 삽입과 삭제를 매우 자주 해야 하는 상황이라면, 배열은 조금 비효율적
* 문제 해결을 위해 연결리스트가 등장함!

> 연결리스트는 삽입과 삭제에 최적화된 구조 O(1)

* 연결리스트는 탐색O(N)은 느림 : head에서 tail까지 일일이 확인해야 함
* 삽입과 삭제 연산은 매우 빠름 (둘 모두 O(1)) : 인접한 곳에서 선을 끊고 연결해주는 작업만 하면 됨
* 일방향으로만 진행되기 때문에 뒤로 돌아갈 수 없는 구조

> 노드 (node) - 정보를 담는 하나의 창구 역할

* 하나의 노드 = 데이터 + 다른 노드로 이동하는 경로
* 여러 개의 노드 = 연결리스트

> 단일 = 연결 방향이 단방향 = 일방통행 구조




* 각 노드별로 data와 next값을 가지고 있는데, Data는 값을, Next는 그 다음 노드의 위치를 가리키는 역할



* 다음 노드가 없기 때문에 Next에는 기본적으로 null(비어있다는 뜻)

> node의 head, tail


* 리스트의 첫번째 값의 위치를 반드시 명시해야하며, 우리는 이것을 head라고 호칭



* 종료지점을 명시해놓으면 탐색할 때 추가적인 처리 없이 현재 방문한 노드가 종료 지점인지 판단하는 과정만 거쳐서 탐색을 종료
