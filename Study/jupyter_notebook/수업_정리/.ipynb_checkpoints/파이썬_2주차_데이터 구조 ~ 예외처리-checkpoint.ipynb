### 1) s.find(x) 와 s.index(x)


```python
# find(x) 함수 사용 
case1 = 'apple'.find('p')  # 1 반환
case2 = 'apple'.find('k')  # -1 반환

print(case1)
print(case2)
```

```python
# index(x) 함수 사용
case1 = 'apple'.index('p')   # 1 반환
case2 = 'apple'.index('k')   # 오류 반환

print(case1)
#print(case2)
```

### 2) s.istitle()


```python
'Title Title!'.istitle()  # True 반환
```

### 3) 숫자 판별 메소드

```python
# 해당 문자열이 0~9까지의 수로 이루어진 것인지 검사합니다.
# 음수로 표현되는 '-'가 붙어있을 시에는 False를 리턴합니다.
case1 = '486'.isdecimal()  # 


# isdecimal()함수의 특징을 모두 가지고 있으며, 해당 문자열이 '숫자'로 이루어져 있는지 검사합니다.  
# 특수기호가 숫자인거 까지 숫자로 포함합니다.(ex 지수)
# 음수로 표현되는 '-'가 붙어있을 시에는 False를 리턴합니다.
case2 = '486'.isdigit()


# isdigit()함수의 특징을 가지고 있으며, 숫자값으로 사용된 문자이면 어떤 방식이든지 숫자로 인정됩니다.
# 특수기호 또한 숫자값으로 사용된다면 숫자로 포함합니다 (ex. 분수)
# 음수로 표현되는 '-'가 붙어있을 시에는 False를 리턴합니다.
case3 = '486'.isnumeric()

print(case1, case2, case3)
```

## 값 추가 및 삭제
### 1) L.append(x)와 L.extend(m)의 차이





```python
cafe = ['star', 'tomn', 'holl']
print(cafe)
cafe.append(['coffee', 'aaae'])
print(cafe)
```

```python
# extend는 원소를 풀어서 추가함 (iterable한 항목)

cafe = ['star', 'tomn', 'holl']
print(cafe)
cafe.extend(['coffee', 'aaae'])
print(cafe)
```

```python
cafe = ['star', 'tomn', 'holl']
print(cafe)
cafe.extend(['coffee'])
print(cafe)
```

```python
cafe = ['star', 'tomn', 'holl']
print(cafe)
cafe += (['coffee'])
print(cafe)
cafe.extend('coffee')
print(cafe)
```

### 6) L.sort()와 sorted


```python
# L.sort()는 원본 리스트를 정렬하고 None을 반환한다
# sort(L)는 원본 리스트와 정렬된 리스트를 반환한다

cafe = ['star', 'tomn', 'holl']
cafe_result = cafe.sort()
print(cafe, cafe_result)
```

```python
cafe = ['star', 'tomn', 'holl']
cafe_result = sorted(cafe)
print(cafe, cafe_result)
```
