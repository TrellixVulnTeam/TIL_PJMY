## `all()` 직접 구현하기

> `all()`은 인자로 받는 iterable(range, list)의 모든 요소가 참이거나 비어있으면 True를 반환합니다. 
>
> 파이썬 내장 함수 `all()`을 직접 구현한 `my_all()`을 작성하시오.


예시)

```python
my_all([]) #=> True
my_all([1, 2, 5, '6']) #=> True
my_all([[], 2, 5, '6']) #=> False
```

````python
# 아래에 코드를 작성하시오.

#def my_all(elements):
# print(bool([]))
# my_all = [1, [1], 2, 5, '6']
    
'''
1. 원소가 1개 이하일 때
1) 빈 리스트일 때([])
2) 요소가 참일 때
3) 요소가 거짓일 때

2. 원소가 2개 이상일 때
1) 하나라도 거짓일 때
2) 모든 요소가 참일 때
'''
def my_all(elements):
    n = len(elements)
    if n <= 1:
        if elements == []:
            return True
        elif bool(elements[0]) == True:
            return True
        elif bool(elements[0]) == False:
            return False
    else:
        for idx1 in elements:
            if bool(idx1) == True:
                continue
            else:    
                break
        # print(idx1)
        return bool(idx1)
        
print(my_all([]))
print(my_all([1, 2, 5, '6']))
print(my_all([[], 2, 5, '6']))
# print(my_all([[]]))
````

## `any()` 직접 구현하기

```python
my_any([1, 2, 5, '6']) #=> True

my_any([0]) #=> False
```

````python
print(my_any([1, 2, 5, '6']))
print(my_any([[], 2, 5, '6']))
print(my_any([0]))
print(any([1, 2, 5, '6']), any([[], 2, 5, '6']), any([0]))
````
