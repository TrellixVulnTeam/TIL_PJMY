```python
# L.sort()는 원본 리스트를 정렬하고 None을 반환한다
# sort(L)는 원본 리스트와 정렬된 리스트를 반환한다

cafe = ['star', 'tomn', 'holl']
cafe_result = cafe.sort()
print(cafe, cafe_result)
```

```python
cafe = ['star', 'tomn', 'holl']
cafe_result = sorted(cafe)
print(cafe, cafe_result)
```

```python
# .append()는 리스트에 값을 그대로 추가해준다


cafe = ['star', 'tomn', 'holl'] 
print(cafe)   # ['star', 'tomn', 'holl']

cafe.append(['coffee', 'aaae']) 
print(cafe)   # ['star', 'tomn', 'holl', ['coffee', 'aaae']]

cafe.append([6, 7, 8])
print(cafe)   # ['star', 'tomn', 'holl', ['coffee', 'aaae'], [6, 7, 8]]
```

```python
# .extend()는 리스트에 값의 원소들의 반복을 풀어서 추가해준다.  

cafe = ['star', 'tomn', 'holl'] 
print(cafe)   # ['star', 'tomn', 'holl']

cafe.extend(['coffee', 'aaae']) 
print(cafe)   # ['star', 'tomn', 'holl', 'coffee', 'aaae']

cafe.extend([6, 7, 8])
print(cafe)   # ['star', 'tomn', 'holl', 'coffee', 'aaae', 6, 7, 8]
```

```python
bloods = ['A', 'B', 'A', 'O', 'AB', 'AB', 'O', 'A', 'B', 'O', 'B', 'AB']

# 딕셔너리에 키가 없으면 for문으로 자동 생성이 되나보다.
blood_dict = {}
for blood in bloods:
    # blood_dict에 키가 존재 한다면, 
    if blood_dict.get(blood):
        print(blood_dict)
        blood_dict[blood] += 1
    else:
        blood_dict[blood] = 1

print(blood_dict)
```

```python
# 1. 문자열을 순회하면서 문자열의 첫 등장 여부를 따지고

# 2. 첫 등장이 아니라면 리스트에 원소를 넣어 준다.

# 3. 그 리스트를 출력한다.
def duplicated_letters(words):
    res = []
    curchr =''

    for i in words:
        if i in curchr:
            if i in res:
                continue
            else:
                res+= i
        else:
            curchr += i
    
    return res

# duplicated_letters('apple')
duplicated_letters('banana')
```

```python
def low_and_up(words):

    new_words = ''

    for i in range(0, len(words)):
        if i % 2 == 0:
            new_words += words[i].lower()
        else:
            new_words += words[i].upper()
    return new_words

# low_and_up('banana')
```
