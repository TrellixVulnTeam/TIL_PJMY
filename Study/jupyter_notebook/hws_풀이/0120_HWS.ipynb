```python
# secret_number = 'tom'

def get_secret_number(secret_number):
    total = 0
    for i in secret_number:
        # print(ord(i))
        total += ord(i)
    # print(total)
    return total

# get_secret_number('tom')
```

```python
def get_strong_word(*words):
    word1 = 0
    word2 = 0
    for i in words:
        #print(i)
        if i == words[0]:
            for w1 in i:
                word1 += ord(w1)
    
        elif i == words[1]:
            for w2 in i:
                word2 += ord(w2)

    # print(word1, word2)
    if word1 > word2:
        return words[0]
    else:
        return words[1]

# get_strong_word('z', 'a')
# get_strong_word('tom', 'john')
```
