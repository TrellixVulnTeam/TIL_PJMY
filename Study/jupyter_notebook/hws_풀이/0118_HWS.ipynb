```python
scores = [80, 89, 99, 83]

# 총점을 구하는 total_score 변수를 선언 및 초기화
total_score = 0

# for문을 사용해 total_score에 scores 리스트의 각 원소 score를 더함 = 총점 구함
for score in scores:
    total_score += score

# total_score 을 len(scores)으로 나눠 총점 / 총 개수 = 평균을 구함
print(total_score / len(scores))

# print((80 + 89 + 99 + 83) / 4)
```

```python
numbers = [85, 72, 38, 80, 69, 65, 68, 96, 22, 49, 67,
          51, 61, 63, 87, 66, 24, 80, 83, 71, 60, 64,
          52, 90, 60, 49, 31, 23, 99, 94, 11, 25, 24]

'''
중간값 = median 
자료를 정렬한 후 위치상으로 중앙의 위치하는 값이 몇인지 구하는 것
전체 자료가 n개라고 할때
자료가 홀수일 때는 n/2 +1에 위치한 값이 중간 값 
짝수일 때는 (n/2 + n/2+1에 위치한 값)/2 가 중간 값
'''
# 1. numbers 를 소팅시킨다.
# 버블 정렬 : https://blog.naver.com/jiyong615/222032133948 참조

for i in range(len(numbers)-1): # 1부터 len(numbers)-1 까지 반복
    for j in range(len(numbers)-i-1): # i부터 len(numbers)-1 까지 반복
        if numbers[j] > numbers[j+1]: # numbers[0] > numbers[1]
            numbers[j], numbers[j+1] = numbers[j+1], numbers[j]

            
# 2. median 인덱스 위치를 구해놓음
m = len(numbers) // 2


# 3. 자료의 개수를 짝수와 홀수로 나눠서 판별한다. 
# 단, 인덱스가 0부터 시작하니까 -1을 해야함
if len(numbers) % 2 == 0:
    print((numbers[m-1] + numbers[m]) / 2)
else:
    print(numbers[m])       
```
