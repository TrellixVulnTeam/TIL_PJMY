## 득표수 구하기

> 주어진 리스트는 반장 선거 투표 결과이다. 이영희의 총 득표수를 출력하시오.

---

```
[출력 예시]
4
```

````python
students = ['이영희', '김철수', '이영희', '조민지', '김철수', '조민지', '이영희', '이영희']

# 아래에 코드를 작성하시오.


# print(set(students)) : 중복 제거해보니 {'김철수', '이영희', '조민지'}가 나옴


# 횟수 변수 cnt 생성 및 초기화
cnt = 0

# students 리스트 안의 원소를 하나씩 출력하는 반복문을 만듦
for student in students:
    
    # 이영희라면, cnt를 1씩 더해준다.
    if student == '이영희':
        cnt += 1

print(cnt)
````

## 최댓값과 등장 횟수 구하기

> 주어진 리스트의 요소 중에서 최댓값과 등장 횟수를 출력하시오.

---

```
[출력 예시]
22 3
```

````python
numbers = [7, 10, 22, 7, 22, 22]

min_num = numbers[0]
cnt = 0
n = len(numbers)

for i in numbers:
    if min_num < i:
        min_num = i

for j in range(0, n):
    if min_num == numbers[j]:
        cnt += 1

print(min_num, cnt)
````

````python
numbers = [7, 10, 22, 7, 22, 22]

# 아래에 코드를 작성하시오.
numbers.sort(reverse = True)

print(numbers[0], len(set(numbers)))
````

## 5의 개수 구하기

> 주어진 리스트의 요소 중에서 5의 개수를 출력하시오.
---
```
[출력 예시]
3
```

````python
numbers = [7, 17, 10, 5, 4, 3, 17, 5, 2, 5]

# 아래에 코드를 작성하시오.

cnt = 0

# numbers 리스트 안의 원소를 하나씩 출력하는 반복문을 만듦
for number in numbers:
    
    # 5라면, cnt를 1씩 더해준다.
    if number == 5:
        cnt += 1

print(cnt)
````

## 'a'가 싫어


> 입력으로 짧은 영단어 word가 주어질 때, 해당 단어에서 'a'를 모두 제거한 결과를 출력하시오.

---
```
[입력 예시]
apple

[출력 예시]
pple
```

````python
word = input()

# 아래에 코드를 작성하시오.


for idx in range(0, len(word)):
    
    # 문자열 처리 replace() 메서드 : 문자열 내 원하는 문자 대치
    # string.replace('바꿀 문자', '이 걸로 바꿀 거임!')
    word = word.replace('a','')

print(word)
````

## 단어 뒤집기

> 입력으로 짧은 영어단어 word가 주어질 때, 해당 단어를 역순으로 뒤집은 결과를 출력하시오.

---
```
[입력 예시]
apple

[출력 예시]
elppa
```

````python
word = input()

# 아래에 코드를 작성하시오.

# 빈 문자열 생성 및 초기화
word_reverse = ''



# apple일 경우 idx : 0,1,2,3,4를 가지고 있고 range(4, -1, -1)과 같으므로 역순으로 출력하는 for문 생성
for i in range(len(word)-1, -1, -1):
    # 역순으로 문자열에 담음
    word_reverse += word[i]

print(word_reverse)
#print(type(word_reverse))

''' [start:end:step]
print(word[::-1])

for i in range(1,6):
        print(word[5-i], end="")'''
````
