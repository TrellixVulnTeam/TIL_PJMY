```python
def queadratic_formula(a,b,c):
    D = (b**2) - (4*a*c)
    # D가 0<0, D==0, D>0일 때 각각 해가 달라짐
    
    if D>0:
        x1 = (-b + D**0.5) / (2*a)
        x2 = (-b - D**0.5) / (2*a)
        print(a,"x^2 + ",b,"x + ",c,"의 해는")
        print("{} 또는 {} 입니다.".format(x1,x2))
    elif D==0:
        x = -b / 2*a
        print("중근입니다. 해는 {}입니다.". format(x))
    else:
        print("허근입니다.")

        
a = 3
b = 4
c = 5
queadratic_formula(a,b,c)
```
