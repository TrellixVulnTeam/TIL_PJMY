```python
# count() 안씀
def count_vowels(words):
    
    mo_list = ['a', 'e', 'i', 'o', 'u']
    cnt = 0
    
    for i in mo_list:
        for word in words:
            if word == i:
                cnt+= 1
    
    return cnt

count_vowels('apple')
#count_vowels('banana')
```

```python
def only_square_area(*two_list):
    
    sa_list = []

    # 1. 각 각 리스트를 쪼갠다.
    for i in two_list[0]:
        for j in two_list[1]:
            # 2. 첫 리스트와 두번 째 리스트의 원소를 비교해서 같은 걸 찾아낸다.
            if i == j:
                sa_list.append(i * j)
    
    # 3. 정사각형의 넓이를 담은 list를 반환한다.
    return sa_list
    
    
    
    
only_square_area([32, 55, 63], [13, 32, 40, 55])
```

```python
def only_square_area(A, B):
    area = []
    for a in A:
        for b in B:
            if a == b:
                area.append( a * b ) 
    return area
```

```python
def get_dict_avg(subjects):

    total = 0
    num = 0
    for i in subjects.values():
        total += i 
        num = len(subjects.values())
    return total / num
    
get_dict_avg({'python' :80, 'algorithm': 90, 'django' : 89, 'web' : 83})
```

```python
def get_dict_avg(my_dict):
    total = 0 
    for key in my_dict:
        total += my_dict[key]
    my_average = total / len(my_dict)
    return my_average

def get_dict_avg(grade):
    sum_ = 0
    for value in grade.values():
        sum_ += value
    
    avg = sum_/len(grade)
    return avg
```

```python
def count_blood(blood):
    type = ['A', 'B', 'O', 'AB']
    count_blood =[]
    dict_ = {}
    for t in type:
        count_blood.append(blood.count(t))
    for i in range(len(type)):
        dict_.update({type[i] : count_blood[i]})
    return dict_
```

```python
def count_blood(blood) :
    li = []
    
    for i in blood :
        li.append(i)
        
    idx = list(set(li))
    
    cnt = {}
    for i in idx :
        cnt[i] = blood.count(i)
        
    return cnt
```
