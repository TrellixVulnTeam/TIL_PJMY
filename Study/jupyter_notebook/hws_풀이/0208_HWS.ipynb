# Web 04. Grid System & Responsive Web (HW)

**`flex-direction`** [CSS](https://developer.mozilla.org/ko/docs/Web/CSS) 속성은 flex container 내의 아이템을 배치할 때 사용할 주축 및 방향(정방향, 역방향)을 지정합니다.



1) row, 특징: flex container의 주축이 글의 작성 방향과 동일합니다. 주축의 시작점과 끝점이 콘텐츠 방향과 동일합니다.



2) row-reverse, 특징:  `row`와 동일하게 동작하지만 시작점과 끝점이 반대에 위치합니다.



3) column, 특징: flex container의 주축이 블록 축과 동일합니다. 주축의 시작점과 끝점이, 글 작성 모드의 이전 지점 및 이후 지점과 동일합니다.



4) column-reverse, 특징: `column`과 동일하게 동작하지만 시작점과 끝점이 반대에 위치합니다.





1) .flex-row



2) .flex-row-reverse



3) .flex-column



4) .flex-column-reverse

1) stretch, 특징: Flex 항목은 너비 및 높이 제약 조건을 준수하면서 항목 여백 상자의 교차 크기가 선과 동일하도록 늘어납니다.



2) flex-start, 특징: Flex 항목의 교차 시작 여백 가장자리는 선의 교차 시작 가장자리와 함께 플러시됩니다.



3) flex-end, 특징: Flex 항목의 교차 끝 여백 가장자리는 선의 교차 끝 가장자리와 플러시됩니다.



4) center, 특징: Flex 항목의 여백 상자는 교차 축의 선 중앙에 있습니다. 항목의 교차 크기가 플렉스 컨테이너보다 크면 양방향으로 동일하게 오버플로됩니다.





(1) flex-direction, flex-wrap

a) container



b) row



c) xs, sm, md, lg, xl, xxl (6개의 grid system breakpoints)



d) 12이하의 숫자 (column의 수)
