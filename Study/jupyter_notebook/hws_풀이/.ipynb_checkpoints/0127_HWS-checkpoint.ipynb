# Python 08. 객체 지향 프로그래밍 (HW)

```python
class Circle:
    pi = 3.14
    
    def __init__(self, r, x, y):
        self.r = r
        self.x = x
        self.y = y
        
    def area(self):
        return Circle.pi * self.r * self.r
    
    def circumference(self):
        return 2 * Circle.pi * self.r
    
    def center(self):
        return (self.x, self.y)

circle_instance = Circle(3, 2, 4)
print(circle_instance.area(), circle_instance.circumference())
```

```python
class Animal:
    def __init__(self, name):
        self.name = name
    
    def walk(self):
        print(f'{self.name}! 걷는다!')
    
    def eat(self):
        print(f'{self.name}! 먹는다!')
              

class Dog(Animal):
    def __init__(self, name):
        self.name = name
        
    def bark(self):
        print(f'{self.name}! 짖는다!')
        
class Bird(Animal):
    def __init__(self, name):
        self.name = name
        
    def fly(self):
        print(f'{self.name}! 푸드덕!')
        
              
dog = Dog('멍멍이')
dog.walk()
dog.bark()


bird = Bird('구구')
bird.walk()
bird.eat()
bird.fly()
```

# Python 07. 객체 지향 프로그래밍 (WS)

```python
class Point:
    
    # 생성자 메서드 정의
    def __init__(self, x, y):
        self.x = x
        self.y = y
        
'''
class Rectangle 에서 파라미터 p1, p2를 처리해줄때
(p1.x, p1.y)처럼 넘겨주는 것을 몰라서 헤맨 것 같다.......
'''

class Rectangle:
    def __init__(self, p1, p2):
        self.p1 = (p1.x, p1.y)
        self.p2 = (p2.x, p2.y)
        
    #def get_area:
    
    def get_area(self):
        return abs((self.p1[0]-self.p2[0]) * (self.p1[1]-self.p2[1]))
    def get_perimeter(self):
        return 2 * (abs(self.p1[0]-self.p2[0]) + abs(self.p1[1]-self.p2[1]))
    def is_square(self):
        return abs(self.p1[0]-self.p2[0]) == abs(self.p1[1]-self.p2[1])

p1 = Point(1, 3)
p2 = Point(3, 1)
r1 = Rectangle(p1, p2)
print(r1.get_area())
print(r1.get_perimeter())
print(r1.is_square())
p3 = Point(3, 7)
p4 = Point(6, 4)
r2 = Rectangle(p3, p4)
print(r2.get_area())
print(r2.get_perimeter())
print(r2.is_square())
```
