에러와 예외
* 에러 : 문법 오류 (외우지 말기 나올때 마다 캐치하기)
* 예외 : 예상치 못한 상황을 내가 핸들링할 수 있음
ex) nameerror, typeerror

계층 구조에서는 아래로 내려갈 수 록 작은 에러임!

* 문제가 발생될 것 같은 에러를 try에 다 넣고,
except에서 처리해준다!

```python
try
    내가 실행시키고 싶은 코드
    
except 이렇게 해서 터지면:
    예외처리 명령문
    
else: (except가 발생하지 않으면)
    실행하는 코드
    
finally:
    터졌건, 잘 됐든 간에 이거 실행해!
```
