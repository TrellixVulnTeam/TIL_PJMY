# 파이썬은 모두 객체(object)로 이뤄져 있다. = 모든 게 클래스다.

* 객체(things) = 상태(속성=value, attribute) + 동작(method)
    * 함수가 메소드보다 더 큰 개념
    * 클래스가 정의되면 클래스 내부에 정의된 함수를 메소드라고 한다.



* 인스턴스 = 사례 = 붕어빵 = 만들어진 객체
    * 클래스로 인해 객체가 실제로 만들어지면 인스턴스라고 함
    * 메모리에 실제로 만들어진 객체를 인스턴스라고 함
 
    
    
* 객체 지향 프로그래밍 : object(속성 + 메소드)를 묶자!
    * 절차 지향 프로그래밍과는 아예 다른 것인가? 경쟁구도인가? NO! 객체 지향 프로그래밍이여도 절차 지향으로 프로그램을 실행하긴 한다. 
    * 절차 지향 프로그래밍의 단점을 해소
    * 절차 지향 프로그래밍보다 속성 + 메소드를 묶어서 유지보수가 편리해졌고, 협업도 좋아짐 그래서oop가 중심이 됨.
    
    
* 클래스명은 파스칼케이스로 (대문자 ex: Person)
* 클래스를 만든다는 것은 새로운 데이터 타입을 만드는 것과 같음
