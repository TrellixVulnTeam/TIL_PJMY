# Python 07. 객체 지향 프로그래밍 (HW)

```python
print(type([1,2,3]))
print(type('sss'))
print(type((3,4)))
print(type(True))
print(type(1))
```

int, str, tuple, bool, list, float, dict

`--init__` : 생성자 메소드, 데이터를 초기화 하는 등의 목적으로 사용됨 
`__del__` : 객체가 소멸될 때 호출됨
`__str__` : 해당 객체의 출력 형태를 지정, 프린트 함수를 호출할 때 자동 호출
`__repr__` : 객체의 출력 형태를 지정, 반환값을 eval함수에 사용할 수 있음(representation의 약자; 표현하다)
