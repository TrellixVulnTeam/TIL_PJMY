## 3-3) 1234 \n 5678을 입력받아 [[1, 2, 3, 4], [5, 6, 7, 8]]로 출력하는 것

예시)

입력 : 1234

      5678

출력 : [[1, 2, 3, 4], [5, 6, 7, 8]]

```python
# 여기서 _는 for문안에서 사용되지 않는 값을 말한다. 
# 즉, 그냥 2번 반복할 것임~


# 빈 객체 생성 및 초기화
matrix = []

for _ in range(2):
    row = list(map(int, input()))
    matrix.append(row)
print(matrix)
```

## 4) 5 * 5의 matrix 만들기 (모든 값은 0)


```python
# 교수님 코드

zeros = []
for i in range(5):
    row = [0] * 5
    zeros.append(row)

# print(zeros)

for row in zeros:
    print(row)
```

```python
# 내가 생각한 코드

N = int(input())
matrix = [0] * N

for idx in range(N): # 0~4
    print(matrix)
```

# 리스트의 얕은 복사 & 깊은 복사

## 1) 복사 방법 : 할당

```python
# 값을 =으로 할당해주는 것은 당연히 얕은 복사가 일어남.
# 나도 문제를 풀면서 실수 했었음...조심하자~~!!

lst = [1, 2, 3, 4]
temp = []

temp.append(lst)
# print(temp) [[1, 2, 3, 4]]

lst.pop()
lst.insert(0, 5) # 여기서 lst는 [5, 1, 2, 3]이 된다.

print(temp) # [[5, 1, 2, 3]]
```

## 2) 1차원 리스트를 깊은 복사하는 법은 슬라이싱 [:]
이 방법은 []내 원소만 복사함



```python
lst = [1, 2, 3, 4]
temp = []

temp.append(lst[:])
# print(temp) [[1, 2, 3, 4]]

lst.pop()
lst.insert(0, 5) # 여기서 lst는 [5, 1, 2, 3]이 된다.

print(temp)
```

## 빈 매트릭스 만들기를 만들 때 주의할 점 : 얕은 복사



```python
zeros = [[0] * 5] * 5

zeros[0][0] = 99

print(zeros) # 띠용 ... 
```
